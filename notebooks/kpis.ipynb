```python
def get_kpis(dff) -> dict:
    trips: pd.DataFrame = dff

    total_trips = trips["trip_count"].sum()
    kpi_total_trips = total_trips

    aggregated = trips.groupby("customer_type")["trip_count"].sum()
    if total_trips == 0:
        member_pct = 0
        casual_pct = 0
    else:
        member_pct = (aggregated.get("member", 0) / total_trips * 100).round(2)
        casual_pct = (aggregated.get("casual", 0) / total_trips * 100).round(2)
    kpi_split_pct = f"{member_pct}/{casual_pct}"

    kpi_avg_time = float(trips["trip_time"].agg("mean").round(2))

    active_station = (
        trips[trips["station_role"] == "start"]
        .sort_values("s_trip_count", ascending=False)
        .iloc[0]
    )
    kpi_active_station = active_station["station_name"]

    hour_agg = trips.groupby("hour")["b_trip_count"].sum()
    kpi_busiest_hour = str(hour_agg.idxmax())

    return {
        "Total Trips": kpi_total_trips,
        "Member vs Casual split %": kpi_split_pct,
        "Avg Trip Duration": kpi_avg_time,
        "Most Active Station": kpi_active_station,
        "Busiest Hour": kpi_busiest_hour,
    }


display(get_kpis(df))
display(get_kpis(df_zero))
```
